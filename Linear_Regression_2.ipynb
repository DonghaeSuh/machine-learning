```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
life = pd.read_csv(data_loc +'life_expectancy.csv')
life = life[['Life expectancy','Alcohol','Percentage expenditure','Polio','BMI','GDP','Thinness 1-19 years']]
print(life.head())
print(life.shape)
```

```python
print(life.isnull())
```

```python
print(life.isnull().sum(axis=0))

life.dropna(axis=0,how='any',inplace=True)
print(life.shape)
```

```python
sns.set(rc={'figure.figsize':(12,10)})
corr_mat =life.corr().round(2)
sns.heatmap(data=corr_mat, annot=True)
```

```python
sns.pairplot(life[['Life expectancy','Percentage expenditure', 'GDP']])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = life[['Alcohol','Percentage expenditure','Polio','BMI','GDP', 'Thinness 1-19 years']]
y=life['Life expectancy']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lin_model =LinearRegression()
lin_model.fit(X_train,y_train)

y_hat_train = lin_model.predict(X_train)
plt.scatter(y_train, y_hat_train)
xy_range = [40,100]
plt.plot(xy_range,xy_range)
```

```python
y_hat_test = lin_model.predict(X_test)
plt.scatter(y_test,y_hat_test)
xy_range =[40,100]
plt.plot(xy_range,xy_range)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE : ', mean_squared_error(y_test,y_hat_test))
```
